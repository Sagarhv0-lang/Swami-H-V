```python
string = input("enter a string:")
count=0
for char in string:
    if char.isupper():
        count += 1
    print("number of uppercase letters:",count)
```

```python
def is_prime(num):
    if num<2 :return false
    for i in range (2,int(num*0.5)+1):
        if num%i==0:return
    return True

n=int(input("enter a number to check if its prime:"))
print(f"{n}is prime" if is_prime(n) else f"{n} is not a prime")
```
