```python
#write a decorator that only aallows npositive num from the user

def decor(fun1):
    def inner(num1):
        if num1 < 0:
            print ("Not a positive NO ")
        else:
            print ("The no is positive ")
    return inner

@decor
def input_avg(num1):
    print(num1)
num1=int(input("Enter the NO :"))
input_avg(num1)
```
